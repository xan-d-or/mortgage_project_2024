```python
print('Введите URL по региону Москва с нужной датой из Домклика')
URL = input() #с нужной датой из Домклика
```

```python
# пример для Москвы для проверки правильно введенных данных (самопроверка)
driver = webdriver.Chrome()
driver.get(URL)

time.sleep(5)

cities = driver.find_elements(By.CSS_SELECTOR, 'div.tg-text-regular--light-5b2-6-2-1')
prices = driver.find_elements(By.CSS_SELECTOR, 'div.tg-text-medium--light-54c-6-2-1 span')
month = ''

data = []
for i in range(len(cities)):
    city = cities[i].text
    price = prices[i].text
    data.append([city, price])

sample = pd.DataFrame(data, columns=['Город', 'Стоимость'])
sample 
```

```python
df_old = pd.read_csv('df2.csv') #файл,который обновляем
df_old2 = df_old[['City', 'P_avg_per_m_sec', 
       'month', 'dummy_space',
       'dummy_time', 'year', 't', 'Population', 'Reg_russian', 'Federal_district',
       'lat', 'lon', 'city_id']]

df_old2.loc[df_old2['Reg_russian'].str.contains('Республика Северная Осетия — Алания'), 'Reg_russian'] = 'Республика Северная Осетия'
df_old2.loc[df_old2['Reg_russian'].str.contains('Ненецкий автономный округ'), 'Reg_russian'] = 'Нарьян-Мар'
df_old2.loc[df_old2['Reg_russian'].str.contains('Ханты-Мансийский автономный округ — Югра'), 'Reg_russian'] = 'Ханты-Мансийский автономный округ'
df_old2['city_rk'] = df_old2['City'] + '_' + df_old2['Reg_russian']
df_old2.head()


regions = df_old2['Reg_russian'].unique()
regions = sorted(regions)
regions 
len(regions)
```

```python
driver = webdriver.Chrome()
driver.get(URL)
#time.sleep(5)


data = []


for region in regions:
    city_input = driver.find_element(By.CSS_SELECTOR, 'input.inpt-input-3c4-11-0-10')

    city_input.send_keys(Keys.CONTROL + "a")
    city_input.send_keys(Keys.DELETE)

    city_input.send_keys(region)
    print(f'Ввел регион: {region}')

    time.sleep(5)

    city_input.send_keys(Keys.DOWN)
    city_input.send_keys(Keys.ENTER)
    print(f'Перешел на страницу с результатами для региона: {region}')

    time.sleep(3)

    cities = driver.find_elements(By.CSS_SELECTOR, 'div.tg-text-regular--light-5b2-6-2-1')
    prices = driver.find_elements(By.CSS_SELECTOR, 'div.tg-text-medium--light-54c-6-2-1 span')

    for i in range(len(cities)):
        city = cities[i].text
        price = prices[i].text
        data.append([region, city, price])

    print(f'Собрал данные для региона {region}: {len(cities)} городов')

driver.quit()

df = pd.DataFrame(data, columns=['Reg_russian', 'City', 'P_avg_per_m_sec'])
df
```

```python
# меняем тип данных у цен вторички
df2 = df[df['P_avg_per_m_sec'] != '—']
df2['P_avg_per_m_sec'] = df2['P_avg_per_m_sec'].str.replace('₽', '')
df2['P_avg_per_m_sec'] = df2['P_avg_per_m_sec'].str.replace(' ', '')
df2['P_avg_per_m_sec'] = df2['P_avg_per_m_sec'].astype(float)
df2 = df2[df2['City'] != 'Другие населенные пункты']
```

```python
# проверка, что ничего не пропало
region_city_counts = df2.groupby('Reg_russian')['City'].nunique().reset_index()
region_city_counts.columns = ['Region', 'Unique_Cities_Count']
region_city_counts2 = df_old2.groupby('Reg_russian')['City'].nunique().reset_index()
region_city_counts2.columns = ['Region', 'Unique_Cities_Count']

compare = region_city_counts.merge(region_city_counts2, how='outer', on = 'Region')
compare.columns = ['Region', 'Unique_Cities_Count_new', 'Unique_Cities_Count_old']

compare
compare[compare['Unique_Cities_Count_new'].isna()]
```

```python
# добавляем айди городов
df2 = df2[['City', 'P_avg_per_m_sec', 'Reg_russian']]
df2['city_rk'] = df2['City'] + '_' + df2['Reg_russian']

df_old3 = df_old2[['dummy_space', 'Population', 'Federal_district', 'lat', 'lon',
       'city_id', 'city_rk']]
df_old3 = df_old3.drop_duplicates()

df3 = df2.merge(df_old3, on = 'city_rk', how = 'inner')
df3
```

```python
# дополнительная информация для таблицы
print('Введите номер выбранного месяца')
month = int(input()) #номер месяца 1,2,...,12
df3['month'] = month

print('Введите номер выбранного года')
year = int(input()) #номер года 2024...
df3['year'] = year
df3['dummy_time'] = 1

data = pd.concat([df_old2, df3], ignore_index=True)

data['date'] = data.year.astype(str) +'-'+data.month.apply(lambda x: str(x) if len(str(x)) == 2 else '0' + str(x)) + '-01'
data['date'] = pd.to_datetime(data['date'])

data = data.sort_values('date')
data['t'] = data['date'].dt.to_period('M').astype(str).factorize()[0] + 1

data = data[['City', 'Population', 'P_avg_per_m_sec', 'date', 'month', 'year',
       't', 'dummy_space', 'dummy_time', 'Reg_russian', 'Federal_district', 'lat', 'lon',
       'city_id', 'city_rk']]
data
```

```python
# кринж-фильтрация
x = data[data['date'] >= '2021-01-01']
x = pd.pivot_table(x, index = 'city_id', aggfunc="count").reset_index()

x = x[['city_id', 'City']].rename(columns = {'City': 'count'})
x

x1 = x[x['count'] == 43]
x1
```
