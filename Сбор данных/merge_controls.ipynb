```python
import pandas as pd
import numpy as np
import time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv('arctic1109.csv')
df = df[['city', 'p_avg_per_m_sec', 'month_dt', 't',
       'dummy_space', 'dummy_time', 'reg_russian', 'federal_district', 'lat',
       'lon', 'city_id', 'city_rk']]

```

```python
dict = pd.read_excel('arctic_dictionary.xlsx')
dummy_reg = dict.loc[dict['dummy_space'] == 1, 'reg_russian'].unique()
dict['dummy_reg'] = dict['reg_russian'].apply(lambda x: 1 if x in dummy_reg else 0)
dict = dict[['city_rk', 'city_id', 'dummy_space', 'dfo_flg', 'dummy_reg', 'okato', 'oktmo',
       'municipal_district_name_short', 'municipal_district_name',
       'municipal_district_type', 'municipal_district_center', 'city',
       'reg_russian', 'oktmo_x', 'oktmo_x2', 'oktmo_city',
       'municipal_district_center_lat', 'municipal_district_center_lon', 'lat',
       'lon', 'territory_id', 'region_code']]
dict
```

```python
dict.columns
```

```python

dict1 = dict[['dfo_flg','oktmo', 'city_rk']]
dict.head()
dict1.head()
```

```python


df = df.merge(dict1, on = 'city_rk', how = 'inner')
df['month_dt'] = pd.to_datetime(df['month_dt'])
df['year'] = df['month_dt'].dt.year
#df =df[df['year'].isin([2022,2023])]

df = df[['city', 'p_avg_per_m_sec', 'month_dt', 'year', 't',
       'dummy_space', 'dummy_time', 'dfo_flg', 'dummy_reg', 'reg_russian', 'federal_district', 'lat',
       'lon', 'city_id', 'city_rk', 'oktmo']]
df.head()
```

```python
dem1 = pd.read_csv('data_Y48112014_year2022_112_v20240402.csv', sep = ';')
dem2 = pd.read_csv('data_Y48112014_year2023_112_v20240402.csv', sep = ';') 

dem = pd.concat([dem1, dem2])
dem = dem[['grup_2', 'vozr', 'region_id', 'region_name',
       'mun_district', 'municipality', 'oktmo', 'mun_type',
       'year', 'indicator_value']]
dem = dem[dem['grup_2'] == 'Всего']
dem = dem.drop(columns=['grup_2'])

demvs = dem[dem['vozr'].isin([ 'Всего'])]
demvs = demvs.rename(columns = {'indicator_value': 
                                 'pop_all'})
demvs = demvs.drop(columns=['vozr'])

demtr = dem[dem['vozr'].isin(['Трудоспособный возраст'])]
demtr = demtr.rename(columns = {'indicator_value': 
                                 'pop_trud'})

demtr = demtr.drop(columns=['vozr'])

dem = demvs.merge(demtr, on = ['region_id', 'region_name', 'mun_district',
       'municipality', 'oktmo', 'mun_type', 'year'] )
dem['oktmo'] = dem['oktmo'].astype(float)

df1 = df.merge(dem, on = ['oktmo', 'year'], how='left')
df1
```

```python
len(df1.loc[(df1['dummy_space'] == 1) & (~df1['pop_trud'].isna()), 'city'].unique())
df1.info()
```
