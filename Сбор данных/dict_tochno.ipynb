```python
from thefuzz import fuzz, process

def fuzzy_match_region(source, target_list, threshold=80):
    # Use process.extractOne for fast fuzzy matching
    match, score = process.extractOne(source, target_list, scorer=fuzz.ratio)
    return match if score >= threshold else None
```
