# Список городов воздействия

```python
#Получаю список городов, в которых было воздействие
treatment_cities = list(set(list(df.query('dummy_space == 1')['city_id'])))
```

```python
sub = df[['City', 'city_id']].drop_duplicates()
cities_dict = {sub.iloc[i, 1]:sub.iloc[i, 0] for i in range(len(sub))}
```

```python
treatment_cities_df = pd.DataFrame({'city_id': treatment_cities, 'City':[cities_dict[city] for city in treatment_cities]})
treatment_cities_df.to_csv('treatment_cities1308.csv')
```
